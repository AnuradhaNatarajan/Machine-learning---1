DATA VISUALIZATION

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer=load_breast_cancer()
```

```python
cancer
```

```python
df_cancer=pd.DataFrame(np.c_[cancer['data'],cancer['target']],columns=np.append(cancer['feature_names'],['target']))
```

```python
df_cancer.head()
```

```python
df_cancer.tail()
```

```python
sns.pairplot(df_cancer,vars=['mean radius','mean perimeter','mean smoothness','mean area'],hue='target')
```

```python
sns.scatterplot(x='mean radius',y='mean smoothness',hue='target',data=df_cancer)
```

```python
sns.countplot(df_cancer['target'])
```

MODEL TRAINING


```python
X=df_cancer.drop('target',axis=1)
```

```python
y=df_cancer['target']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report,confusion_matrix
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=5)
```

```python
svc_model=SVC()
```

```python
svc_model.fit(X_train,y_train)
```

MODEL EVALUATION

```python
y_predict=svc_model.predict(X_test)
```

```python
y_predict
```

```python
cm=confusion_matrix(y_test,y_predict)
```

```python
sns.heatmap(cm,annot=True)
```

```python
min_train=X_train.min()
range_train=(X_train-min_train).max()
X_train_scaled=(X_train-min_train)/range_train
```

```python
svc_model.fit(X_train_scaled,y_train)
```

```python
min_test=X_test.min()
range_test=(X_test-min_test).max()
X_test_scaled=(X_test-min_test)/range_test
```

```python
y_predict=svc_model.predict(X_test_scaled)
```

```python
cm=confusion_matrix(y_test,y_predict)
```

```python
X_test_scaled=X_test-X_test.min()/X_test.max()-X_test.min()
```

```python
cm=confusion_matrix(y_test,y_predict)
```

```python
sns.heatmap(cm,annot=True)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={'C':[0.1,1,10,100],'gamma':[1,0.1,0.01,0.001],'kernel':['rbf']}
```

```python
grid=GridSearchCV(SVC(),param_grid,refit=True,verbose=4)
```

```python
grid.fit(X_train_scaled,y_train)
```

```python
grid.best_params_
```

```python
y_new_predict=grid.predict(X_test_scaled)
```

```python
cm=confusion_matrix(y_test,y_new_predict)
```

```python
sns.heatmap(cm,annot=True)
```

```python
print(classification_report(y_test,y_predict))
```
